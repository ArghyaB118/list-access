```python
import random
import matplotlib.pyplot as plt
```

```python
discount = []
for i in range(5):
    discount.append(2)
for i in range(5):
    discount.append(1)
    
print(discount)
plt.plot(discount)
plt.xlim([0, 10]); plt.xlabel("number of days", fontsize=16)
plt.ylim([0, 3]); plt.ylabel("% of market return", fontsize=16)
plt.show()
```

```python
rent = [1.0]
for i in range(10):
    rent.append(rent[-1] - discount[i] / 100 * 1.0)
print(rent)
plt.plot(rent)
plt.xlim([0, 10]); plt.xlabel("number of days", fontsize=16)
plt.ylim([0.8, 1.1]); plt.ylabel("effective rent of the ski", fontsize=16)
plt.show()
```

```python
rent = [1.0]
for i in range(100):
    rand = random.randint(0, 1000)
    rand = rand / 1000
    rent.append(rent[-1] - rand / 100 * 1)

plt.plot(rent)
plt.xlim([0, 100]); plt.xlabel("number of days", fontsize=16)
plt.ylim([0, 1]); plt.ylabel("effective rent of the ski", fontsize=16)
plt.show()
```
